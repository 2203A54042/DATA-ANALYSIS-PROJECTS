```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import accuracy_score, classification_report
for col in df.columns:
    if df[col].dtype == 'datetime64[ns]':
        df[f'{col}_year'] = df[col].dt.year
        df[f'{col}_month'] = df[col].dt.month
        df[f'{col}_day'] = df[col].dt.day
        df[f'{col}_hour'] = df[col].dt.hour
        df[f'{col}_minute'] = df[col].dt.minute
        df = df.drop(columns=[col])
categorical_cols = df.select_dtypes(include=['object']).columns
label_encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
df = df.apply(pd.to_numeric, errors='coerce')
y_class = df['ride_status']
X_class = df.drop(columns=['ride_status'])
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_class, y_class, test_size=0.2, random_state=42)


logreg = LogisticRegression(max_iter=500)
logreg.fit(X_train_c, y_train_c)
y_pred_logreg = logreg.predict(X_test_c)

rf_clf = RandomForestRegressor(n_estimators=100, random_state=42)
rf_clf.fit(X_train_c, y_train_c)
y_pred_rf_clf = rf_clf.predict(X_test_c)

xgb = XGBRegressor(use_label_encoder=False, eval_metric='mlogloss')
xgb.fit(X_train_c, y_train_c)
y_pred_xgb = xgb.predict(X_test_c)

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Evaluate Regression Models
def evaluate_regression(y_test, y_pred, model_name):
    print(f" {model_name} Results:")
    print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, y_pred))
    print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))
    print("R-squared (RÂ²):", r2_score(y_test, y_pred))
    print("-" * 30)

evaluate_regression(y_test_c, y_pred_rf_clf, "Random Forest Regressor")
evaluate_regression(y_test_c, y_pred_xgb, "XGBoost Regressor")
evaluate_regression(y_test_c, y_pred_logreg, "Logistic Regression")
print()
```

```python
from sklearn.model_selection import cross_val_score, KFold
from xgboost import XGBClassifier

kf = KFold(n_splits=5, shuffle=True, random_state=42)
xgb_clf = XGBClassifier(eval_metric='mlogloss')
scores = cross_val_score(xgb_clf, X_class, y_class, cv=kf, scoring='accuracy')
print(f"XGBoost Classifier CV Accuracy: {np.mean(scores)}")
```
